```python
import os
import json
from dotenv import dotenv_values

def load_env_with_json(path=".env"):
    """
    Đọc file .env và tự động json.loads các giá trị JSON.
    Trả về dict Python.
    """
    env_data = dotenv_values(path)
    parsed = {}

    for key, value in env_data.items():
        if value is None:
            continue
        # Bỏ dấu nháy ngoài nếu có
        clean_value = value.strip()
        # if (clean_value.startswith("'") and clean_value.endswith("'")) or \
        #    (clean_value.startswith('"') and clean_value.endswith('"')):
        #     clean_value = clean_value[1:-1]

        # Thử parse JSON
        try:
            parsed[key] = json.loads(clean_value)
        except json.JSONDecodeError:
            parsed[key] = clean_value

    return parsed

config = load_env_with_json(".env")

# In ra các cấu hình
print("APP_ENV:", config.get("APP_ENV"))
print("\nDEV_OBSERVER_CONFIG:")
print((config.get("DEV_OBSERVER_CONFIG")))

```
